```python
def most_successful(df,sport):
    temp_df = df.dropna(subset=['Medal'])

    if sport != 'Overall':
        temp_df = temp_df[temp_df['Sport']==sport]

    x = temp_df['Name'].value_counts().reset_index().head(15).merge(df, left_on='Name', right_on='Name', how='left')[['Name', 'Medal', 'Sport', 'region']].drop_duplicates('Name')
    x.rename(columns= {'Name': 'Name', 'Medal': 'Medals'}, inplace=True)
    return x 
```

```python
most_successful(df, 'Gymnastics')
```
